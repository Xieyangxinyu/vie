```python
import pandas as pd
import matplotlib.pyplot as plt

def get_box_plot(filename = 'abalone_4177n_8p_10k_15r'):
    data = []

    for resample in range(15):
        df = pd.read_csv(f"real_data_results/{filename}_{resample}.csv")
        df.drop(columns=["fold"], inplace=True)
        data.append(df)
    # compute mean of all resamples and create new dataframe
    summary = []

    for resample in range(15):
        summary.append(data[resample].mean(axis=0))

    summary = pd.DataFrame(summary)

    # drop the column `mean`
    summary.drop(columns=["mean"], inplace=True)

    # change the column names by replacing string `save` with `SAVE`, `sir` with `SIR`, and `rf` with `RF`, `mft` with `MFT` and `mf` with `MF`
    summary.columns = summary.columns.str.replace("save", "SAVE")
    summary.columns = summary.columns.str.replace("sir", "SIR")
    summary.columns = summary.columns.str.replace("rf", "RF")
    summary.columns = summary.columns.str.replace("mft", "MFT")
    summary.columns = summary.columns.str.replace("mf", "MF")


    fig, ax = plt.subplots(figsize=(10, 10))
    summary.boxplot(ax=ax)
    # enlarge the font size of x and y ticks
    plt.xticks(rotation=45)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    # enlarge the font size of x and y labels
    plt.ylabel("Mean Squared Error", fontsize=20)
    plt.xlabel("Model", fontsize=20)
    plt.tight_layout()
    
```

```python
get_box_plot('abalone_4177n_8p_10k_15r')
plt.savefig('real/abalone.png')
```
