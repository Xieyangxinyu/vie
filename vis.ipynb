```python
import pandas as pd
import matplotlib.pyplot as plt

def get_box_plot(filename = 'abalone_4177n_8p_10k_15r'):
    data = []

    for resample in range(15):
        df = pd.read_csv(f"real_data_results/{filename}_{resample}.csv")
        df.drop(columns=["fold"], inplace=True)
        data.append(df)
    # compute mean of all resamples and create new dataframe
    summary = []

    for resample in range(15):
        summary.append(data[resample].mean(axis=0))

    summary = pd.DataFrame(summary)

    fig, ax = plt.subplots(figsize=(10, 10))
    summary.boxplot(ax=ax)
    plt.xticks(rotation=45)
```

```python
get_box_plot('abalone_4177n_8p_10k_15r')
```
